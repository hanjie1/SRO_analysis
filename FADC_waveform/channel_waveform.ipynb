```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
waveforms = np.load('output/fadc_waveforms_141.npy',allow_pickle=True)
fadc_info = np.load('output/fadc_info_141.npy',allow_pickle=True)
```

```python
nevents=50
all_slots=[3,4] # all FADC slots
```

```python
slot=4
chan=10
slot_idx = all_slots.index(slot)

plt.close()
for i in range(nevents):
        
    a_wave = waveforms[i,slot_idx,chan]

    xx = range(len(a_wave))
    if a_wave[0]!=0:
       plt.plot(xx,a_wave)


print(len(xx))
plt.xlabel("samples")
plt.ylabel("ADC value")
plt.title(f"Channel {chan} waveforms")
#plt.legend()
plt.grid(True)
plt.show()
```
