```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
waveforms = np.load('output/fadc_waveforms.npy',allow_pickle=True)
fadc_info = np.load('output/fadc_info.npy',allow_pickle=True)
```

```python
nevents=1000
```

```python
slot=3
chan=2
plt.close()
for i in range(nevents):
    a_slot = fadc_info[i]['slot_id']
    if a_slot!=slot:
        continue
        
    a_wave = fadc_info[i]['waveforms'][chan]
    xx = range(len(a_wave))
    #plt.plot(xx,a_wave, label=f"event {i}")
    if a_wave[0]!=0:
       plt.plot(xx,a_wave)


print(len(xx))
plt.xlabel("samples")
plt.ylabel("ADC value")
plt.title(f"Channel {chan} waveforms")
#plt.legend()
plt.grid(True)
plt.show()
```

```python
chan=4
plt.clf()
for i in range(nevents):
    a_wave = waveforms[i,chan,:20]
    xx = range(len(a_wave))
    #plt.plot(xx,a_wave, label=f"event {i}")
    if a_wave[0]!=0:
       plt.plot(xx,a_wave)


print(len(xx))
plt.xlabel("samples")
plt.ylabel("ADC value")
plt.title(f"Channel {chan} waveforms")
#plt.legend()
plt.grid(True)
plt.show()
```
